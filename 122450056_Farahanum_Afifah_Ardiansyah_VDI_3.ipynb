```python
import numpy as np
import pandas as pd

df1 = pd.read_csv('StudentPerformanceFactors.csv', na_values = ['#NAME'])
data = pd.DataFrame(df1)
data.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Membuat scatter plot dari Hours_Studied, Attendance vs Exam_Score
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Hours_Studied', y='Exam_Score', size='Attendance',
                hue='Attendance', data=data, sizes=(20, 200),
                palette='coolwarm', legend='brief')

# Kustomisasi plot
plt.title('Hubungan Antara Hours Studied, Attendance, and Exam Scores', fontsize=14)
plt.xlabel('Hours Studied', fontsize=12)
plt.ylabel('Exam Score', fontsize=12)
plt.legend(title='Attendance (%)', loc='upper left', bbox_to_anchor=(1, 1), borderaxespad=0)

# Menampilkan plot
plt.tight_layout()
plt.show()
```

## Tujuan
dari visualisasi yang dibuat, memiliki tujuan untuk **mengungkap korelasi** atau **mengidentifikasi hubungan** dalam dataset multivariat. scatterplot memungkinkan pengguna atau user untuk melihat pola bagaimana faktor-faktor ini mempengaruhi kinerja akademik secara visual. serta bertujuan untuk mempermudah interpretasi dengan menggunakan ukuran dan warna untuk menyoroti kehadiran sebagai faktor tambahan yang mempengaruhi hasil

## user (pengguna)
* Guru dan staf pendidikan : membantu untuk memahami pentingnya konsistensi kehadiran dan jam belajar siswa untuk meningkatkan hasil ujian
* Siswa dan Orang tua : Melihat bagaimana jam belajar dan kehadiran memengaruhi hasil akademik dan mengambil langkah yang lebih baik untuk mengoptimalkan waktu belajar

##Fungsi
* Menyajikan cara visual yang interaktif untuk memahami hubungan antar beberapa faktor
* scatterplot membantu untuk **identifikasi pola langsung** antara variabel numerik seperti korelasi positif antara jam belajar dan nilai ujian
